```python

import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import seaborn as sns
```

```python
dataset = pd.read_csv("C:\\Users\\Muzamil Hussain\\Desktop\\deep learning data set\\student_dataset.csv")
dataset.head(5)
```

```python
sns.scatterplot(x="cgpa", y="score", data = dataset, hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha = 1)
pr.fit(x_train, y_train)
```

```python
pr.score(x_train, y_train)*100, pr.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```
